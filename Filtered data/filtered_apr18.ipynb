A week's worth of data (4/14/2014 - 4/20/2014) is still too large to work with. I am filtering my data to one day only. I have chosen Friday 4/18/2014. I am assuming this is a good day to capture both commuter data and Friday nightlife data - a good opportunity for TLC to make the most money.

**Addendum**: After saving these files to .csv and importing to Google Sheets, the data is still too large. I am taking a random sample of the yellow taxi data for use in my analyses.

## Green Taxi

```python
green_taxi = pd.read_csv('filtered_green_taxi.csv')
```

```python
filtered_green_taxi_apr18 = green_taxi[(green_taxi['pickup_datetime'] >= '2014-04-18 00:00:00') & (green_taxi['pickup_datetime'] <= '2014-04-18 23:59:59')]
```

```python
filtered_green_taxi_apr18.head()
```

```python
filtered_green_taxi_apr18.to_csv('filtered_green_taxi_apr18.csv')
```
